```python
import pandas as pd

df = pd.read_csv('/content/canada_per_capita_income.csv')
display(df.head())
```

```python
df.rename(columns={'per capita income (US$)': 'per_capita_income_us'}, inplace=True)
display(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='year', y='per_capita_income_us', data=df)
plt.title('Per Capita Income in Canada Over Years')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df[['year']]
y = df['per_capita_income_us']

model = LinearRegression()
model.fit(X, y)

year_2020 = pd.DataFrame([2020], columns=['year'])
predicted_income_2020 = model.predict(year_2020)

print(f'Predicted per capita income for Canada in 2020: ${predicted_income_2020[0]:,.2f}')
```
