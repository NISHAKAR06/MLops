## Dataset Loading and Serialization Demonstration

```python
import pandas as pd
import pickle
import joblib
import os

file_path = '/content/ai_product_launch_dataset.csv.csv'
df = pd.read_csv(file_path)
print("Dataset loaded successfully:")
display(df.head())
```

### 1. Perform One-Hot Encoding on Categorical Features

```python
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns

categorical_cols = [col for col in categorical_cols if col not in ['crawl_timestamp', 'product_url', 'product_name', 'description', 'image', 'product_category_tree', 'product_specifications']]

print(f"Categorical columns identified for one-hot encoding: {categorical_cols}")

df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

print("\nDataFrame after one-hot encoding:")
display(df_encoded.head())
```

2.PERFORM TRAIN_TEST-SPLIT OPERATION

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import pandas as pd
df_cleaned = df_encoded.dropna(subset=['discounted_price']).copy()

X = df_cleaned.drop('discounted_price', axis=1)
y = df_cleaned['discounted_price']

X = X.select_dtypes(include=['number', 'bool'])

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
X = pd.DataFrame(X_imputed, columns=X.columns, index=X.index)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Original features shape (after cleaning y NaNs and imputing X NaNs): {X.shape}")
print(f"Training features shape: {X_train.shape}")
```

###  Train a Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
import pickle

model = LinearRegression()

model.fit(X_train, y_train)

print("Linear Regression Model trained successfully.")

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE) on test set: {mae:.2f}")
print(f"R-squared (R2) on test set: {r2:.2f}")

pickle_model_filename = 'linear_regression_model.pkl'

with open(pickle_model_filename, 'wb') as f:
    pickle.dump(model, f)
print(f"Linear Regression model successfully pickled and saved to {pickle_model_filename}")
```

### **3**. Store and Load the Model using `pickle`

```python

pickle_model_filename = 'linear_regression_model.pkl'
with open(pickle_model_filename, 'rb') as f:
    loaded_model_pickle = pickle.load(f)
print(f"Linear Regression model successfully loaded from {pickle_model_filename}")

loaded_model_pickle_pred = loaded_model_pickle.predict(X_test.head(1))
print(f"Prediction with loaded pickle model: {loaded_model_pickle_pred[0]:.2f}")
print(f"Actual value for the first test sample: {y_test.head(1).values[0]:.2f}")
```

### 4. Store and Load the Model using `joblib`

```python
import joblib

joblib_model_filename = 'linear_regression_model.joblib'

joblib.dump(model, joblib_model_filename)
print(f"Linear Regression model successfully saved using joblib to {joblib_model_filename}")

loaded_model_joblib = joblib.load(joblib_model_filename)
print(f"Linear Regression model successfully loaded from {joblib_model_filename}")

loaded_model_joblib_pred = loaded_model_joblib.predict(X_test.head(1))
print(f"Prediction with loaded joblib model: {loaded_model_joblib_pred[0]:.2f}")
print(f"Actual value for the first test sample: {y_test.head(1).values[0]:.2f}")
```
